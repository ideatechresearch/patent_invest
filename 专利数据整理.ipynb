```python
def get_hdf_table(path,columns=[]):
    with pd.HDFStore(path) as store:
        if columns:
            dataset=store.get('col_dataset') 
            keys=dataset[dataset.isin(columns)].index #[key for key in store.keys() if store[key].name in columns]
            df=pd.concat([store.get(key) for key in keys], axis=1)
            cols=[col for col in columns if col in df.columns]
            return df[cols]
        
        return pd.concat([store[key] for key in store.keys() if key!='/col_dataset'], axis=1)#全部 
        
def get_hdf_table(path,columns=[]):
    with pd.HDFStore(path,'r') as store:
        if '/col_dataset' in store.keys():
            dataset=store.get('col_dataset')
            if columns:
                keys=dataset[dataset.isin(columns)].index
                df=pd.concat([store.get(key) for key in keys], axis=1)
                cols=[col for col in columns if col in df.columns]#还原顺序
                return df[cols]
            else:
                df=pd.concat([store[key] for key in store.keys() if key!='/col_dataset'], axis=1)#全部 
                cols=[col for col in dataset if col in df.columns]
                return df[cols]

        return pd.concat([store[key] for key in store.keys() if not columns or store.get(key).name in columns], axis=1) 
```
