1. 专精特新

“专精特新”是指“专业化、精细化、特色化、新颖化”的中小企业。这类企业通常专注于某个细分市场，具有较强的创新能力和竞争优势。政府推出“专精特新”企业培育计划，旨在鼓励中小企业深耕某个领域，提升技术水平和市场竞争力。具体特点包括：

    专业化：专注于某个行业或技术领域。
    精细化：在生产、管理和技术方面精细管理。
    特色化：拥有独特产品、技术或服务。
    新颖化：具备创新性，推动技术和产品的持续创新。

专精特新企业一般是中小企业，政策支持方向包括资金、税收优惠、融资等。
2. 高新技术企业

高新技术企业是指符合中国国家高新技术领域，依靠科技进行研发和创新，持续进行技术成果转化并取得市场效益的企业。这些企业往往具备较高的技术水平和科研能力，享有国家的税收优惠和政策扶持。高新技术企业需满足以下条件：

    技术领域：需符合国家重点支持的高新技术领域，如生物、信息、材料等。
    研发投入：企业研发投入占销售收入的比例达到一定要求。
    知识产权：拥有核心自主知识产权。
    成长性：企业有较高的技术产品市场占有率和较强的成长性。

高新技术企业的政策优惠通常包括企业所得税减免、资金补助、创新能力支持等。
3. 税调

“税调”即税务调整，是针对企业税务申报过程中的财务项目进行调整，以符合税法的要求。与其他概念不同，税调本质上是财务管理与税务申报中的一种调整行为，主要用于企业的财务和税收管理。

    内容：涉及会计利润与应纳税所得额的差异调整，税前扣除项目调整，折旧和摊销的调整等。
    对象：适用于所有企业，并不类型企业的政策或资助项目。

4. 创新基金_立项

“创新基金”是中国政府推出的，旨在支持中小企业科技创新的专项基金。企业通过申请创新基金的项目立项，获得资金支持。这种资金支持多用于科研开发、技术创新和科技成果转化。

    支持对象：主要面向科技型中小企业。
    资金用途：用于支持企业创新研发，推动技术进步和产品升级。
    政策目的：通过资金支持激励企业进行技术创新和研发投入。

5. 火炬项目

火炬项目是中国火炬计划的组成部分，是国家支持高新技术产业发展、推进科技成果转化的重要政策。火炬项目主要支持高新技术企业及其高技术产业化项目，目标是促进科技创新和技术转化。

    支持对象：高新技术企业和科研机构。
    项目性质：多为科技成果的转化与产业化。
    政策目的：通过项目申报与资金支持，推动企业和科研单位将创新成果转化为实际生产力。

6. 融资企业

融资企业是指通过金融市场、投资机构或其他融资手段获得资金支持的企业。这类企业可能需要融资来扩大规模、研发新产品或推动市场拓展。融资方式包括股权融资、债务融资、银行贷款、风险投资等。

    融资渠道：企业可以通过银行贷款、股权融资、债券发行、、私募基金等方式获得资金。
    融资目的：通常用于企业扩张、研发投入、并购等目的。

区别总结

    企业类型：专精特新企业和高新技术企业都属于创新型企业，但前者更注重中小企业的专注和特色，后者则要求企业有较强的科技研发能力。融资企业则是融资渠道的需求，不限企业规模或类型。
    政策支持：高新技术企业和专精特新企业都有较多的政策优惠（如税收减免和资金补贴），火炬项目和创新基金更多侧重于项目立项和科技成果转化。税调则是企业财务税务管理中的具体操作。
    


1. 专精特新与专利

“专精特新”企业通常在某一领域内有深入的技术积累和创新能力，专利是它们的关键竞争力之一。这些企业往往在特定技术或工艺领域拥有核心专利或技术，形成差异化的竞争优势。政府的支持政策也鼓励这些企业通过专利保护创新成果，以提高市场竞争力。

    关系：专精特新企业需要通过专利等知识产权手段，保护其在细分市场中的技术创新成果。
    政策支持：在申请专精特新企业时，专利数量和质量（例如发明专利的拥有量）是重要的评估标准之一。

2. 高新技术企业与专利

高新技术企业要求企业具备较强的研发和创新能力，专利在其认定过程中扮演关键角色。高新技术企业必须拥有核心自主知识产权，特别是技术含量高的发明专利、实用新型专利和软件著作权等。拥有高质量的专利数量是衡量企业技术水平的重要标准。

    关系：高新技术企业认证中的核心自主知识产权通常指的是企业拥有的专利，尤其是高价值的发明专利。
    政策支持：通过高新技术企业认证后，企业可以享受税收优惠等多种政策支持，激励其进一步加大研发投入，开发更多专利。

3. 税调与专利

税调（税务调整）主要与企业的财务和税务管理有关，专利在税务处理中的作用体现在研发费用的税前扣除和知识产权的转让收益上。例如，企业在开发专利技术过程中产生的研发费用可以根据税法规定享受税前扣除优惠，专利转让产生的收入也需要进行相关的税务处理。

    关系：企业在进行专利研发、专利转让时，会产生研发费用和收些都需
   在税务调整中进行处理。
    政策支持：专利研发支出可以作为研发费用，在税前进行扣除，降低企业的促进创新活动。

4. 创新基金_立项与专利

创新基金是政府为鼓励中小企业进行科技创新而设立的专项资金，申请创新基金项目时，企业的专利成果往往是评审的重要依据。企业拥有的专利数量和技术水平，能够证明其创新能力，帮助其获得项目资金支持。

6   关系：企业的专利数量和质量往往决定其能否成功立项，获得创新基金的资助。
    政策支持：创新基金支持企业的研发活动，推动专利申请和知识产权保护。

5. 火炬项目与专利

火炬项目是国家支持高新技术产业发展的重要计划，主要针对高新技术企业进行技术成果转化和产业化，专利是其核心资源之一。企业通过火炬项目，可以将研发出的专利技术应用到实际生产中，并实现产业化。

    关系：火炬项目的企业需要具备一定的技术积累和专利成果，专利技术的产业化是项目成功的关键。
    政策支持：火炬项目帮助企业将专利技术成果转化为产品或服务，促进技术落地。

6. 融资企业与专利

对于融资企业来说，专利是其资产价值的重要组成部分，尤其是科技型企业，专利技术的持有和应用往往决定了企业的融资能力和市场估值。投资机构会考量企业的专利数量、质量及其市场前景，作为投资决策的重要依据。

    关系：专利作为无形资产，在企业融资过程中提升企业的市场价值和融资能力。
    政策支持：企业可以通过知识产权质押融资，利用专利等无形资产获取贷款和投资。

总结

这些概念与专利密切相关，尤其是在技术创新、知识产权保护和转化方面。企业通过专利积累，不仅能获得政府的政策支持，还能提升市场竞争力和融资能力。专利在这些企业政策中不仅是技术实力的象征，也是享受政策红利的关键因素。目的与内容：创新基金和火炬项目更侧重于支持企业的研发与科技创新，而税调是税务规范行为，与其他概念有本质不同。

```python
import concurrent.futures
import threading
lock = threading.Lock()
batch_size=10000
output_csv = 'Y:/Downloads/Patent/patent_qymc_2408.csv'
file_exists = pd.io.common.file_exists(output_csv) #os.path.exists(output_csv)
def save_to_csv(df):
    global file_exists
    with lock:
        df.to_csv(output_csv, mode='a', header=not file_exists, index=False)
        file_exists = True 
        
data=pd.read_csv('Y:\Downloads\Patent\invest_all_patent_bridge_2408.csv',index_col=0)
tb_id=data.groupby('table_name')['申请号'].apply(lambda x:x.to_list())#'公开（公告）号'
with concurrent.futures.ThreadPoolExecutor() as executor:
    for table in tb_id.index:
        query = text(f'SELECT {column_name} FROM `{table}` WHERE `申请号` in :ids')
        chunks_list = []
        ids = tuple(tb_id[table])
        
        for batch in chunks(ids, batch_size):
            chunk = pd.read_sql_query(query, con=engine, params={'ids': batch})
            chunk = chunk.sort_values('公开（公告）日').drop_duplicates(subset='公开（公告）号',keep='first') 
            chunks_list.append(chunk)
            print(len(chunks_list),chunk.shape[0])
        
        df_filter = pd.concat(chunks_list, ignore_index=True)
        df_filter['table_name'] = table
        
        # 提交保存操作到线程池
        executor.submit(save_to_csv, df_filter)

        print(f"Finished processing table: {table} with shape: {df_filter.shape}")
        del df_filter
        del chunks_list
```

```python
import re
chinese_to_arabic = {
    '零': '0', '一': '1', '二': '2', '三': '3', '四': '4',
    '五': '5', '六': '6', '七': '7', '八': '8', '九': '9'
}

# 1. 剔除异常符号
def remove_unwanted_symbols(name):
    # 定义允许的字符：中文、英文、数字和常见符号（不想删除的符号）
    allowed_characters = r'[^\w\s\(\)（）【】「」『』]'
    # 删除所有不允许的符号
    return re.sub(allowed_characters, '', name)

# 2. 将各种括号转换成英文括号
def convert_brackets(name):
    # 替换中文和全角括号为标准英文括号
    return name.replace('（', '(').replace('）', ')') \
               .replace('【', '[').replace('】', ']') \
               .replace('「', '[').replace('」', ']') \
               .replace('『', '[').replace('』', ']')

# 3. 将汉字数字替换成阿拉伯数字
def replace_chinese_numbers(name):
    for ch_num, ar_num in chinese_to_arabic.items():
        name = name.replace(ch_num, ar_num)
    return name

# 4. 检查并调整括号匹配
def fix_unbalanced_brackets(name):
    left_count = name.count('(')
    right_count = name.count(')')
    
    # 如果左括号比右括号多，补齐右括号
    if left_count > right_count:
        name += ')' * (left_count - right_count)
    # 如果右括号比左括号多，补齐左括号
    elif right_count > left_count:
        name = '(' * (right_count - left_count) + name
    
    return name

# 5. 移除前面的数字和数字开头的部分
def remove_leading_numbers(name): 
    return re.sub(r'^\d+.*?(?=\D)', '', name).strip()  # re.sub(r'^\d+.*? ', '', name) 移除开头的数字直到第一个非数字字符，并去掉前后空格

    
# 综合清洗函数
def clean_fullname(name):
    if not isinstance(name, str):
        return name
    name = remove_unwanted_symbols(name)  # 剔除异常符号
    name = convert_brackets(name)  # 将括号转换成英文括号
    name = replace_chinese_numbers(name)  # 替换汉字数字为阿拉伯数字
    name = fix_unbalanced_brackets(name)  # 调整不平衡的括号
    name = remove_leading_numbers(name)  # 移除前面的数字
    return name.upper() # 转换为大写并去掉前后空格

# 对 fullname 列应用清洗操作
# data_2['企业全称'] = data_2['fullname'].apply(clean_fullname)
```
