```python
from sklearn.metrics.pairwise import cosine_similarity
doc_lda_w2v=pd.read_parquet(f'data\\documents_vec_stk_list.parquet')
doc_lda_w2v.shape#1985628
```

```python
def year_interval(filter_value,cols):
    filter_value_1 = filter_value.dropna(subset = cols) #filter_value.sort_values(by=['apply_date'],ascending=True)
    return (filter_value_1['申请日'].max()-filter_value_1['申请日'].min()).days//365+1

def process_ipc(filter_patent):
    ipcs_patent = filter_patent['IPC'].str.split(';').explode().str.strip().str.extract('(\w+)/')[0].str[:3]
    return 1.0 - (ipcs_patent.value_counts(normalize=True)**2).sum()

def calculate_metrics(filter_value):
    metrics = {}
    
    # 申请号的唯一数量
    metrics['Application_num'] = filter_value['申请号'].nunique()
    
    # 去重后的 IPC 主分类数量
    metrics['IPC主分类_去重'] = filter_value['IPC主分类'].dropna().map(lambda x: x[:3]).nunique()
    
    # IPC 主分类的平均值
    metrics['IPC主分类_平均'] = metrics['IPC主分类_去重'] / year_interval(filter_value, 'IPC主分类')

    # 计算 IPC 的去重数量和平均值
    ipcs = filter_value['IPC'].str.split(';').explode().str.strip()
    metrics['IPC_个数_去重'] = ipcs.nunique()
    metrics['IPC_个数_平均'] = metrics['IPC_个数_去重'] / year_interval(filter_value, 'IPC')

    # 被引证次数和引证数量
    metrics['被引证次数'] = filter_value['被引证次数'].sum()
    metrics['引证数量'] = filter_value['引证次数'].sum()
    
    # 引证专利的去重数量和平均值
    citing = filter_value['引证专利'].str.split(';').explode().str.strip()
    metrics['引证数量_去重'] = citing.nunique()
    metrics['引证数量_平均'] = metrics['引证数量_去重'] / year_interval(filter_value, '引证专利')

    # 简单同族专利相关统计
    metrics['简单同族个数'] = filter_value['简单同族个数'].sum()
    cognate = filter_value['简单同族'].str.split(';').explode().str.strip()
    metrics['简单同族_去重'] = cognate.nunique()
    metrics['简单同族美日欧_比例'] = cognate.str[:2].isin(['US', 'JP', 'EP']).sum() / cognate.shape[0]

    # 发明人数量
    metrics['发明人数量'] = filter_value['发明人数量'].sum()
    metrics['发明人数量_去重'] = filter_value['发明人'].str.split(';').explode().str.strip().nunique()
    metrics['发明人数量_平均'] = metrics['发明人数量_去重'] / year_interval(filter_value, '发明人')

    # 权利要求数量和引证科技文献数量
    metrics['权利要求数量'] = filter_value['权利要求数量'].sum()
    metrics['引证科技文献数量'] = filter_value['引证科技文献'].count()  # 80%空,格式不统一
    
    # 有效专利数量和有效比例
    metrics['有效数量'] = (filter_value['专利有效性'] == '有效').sum()
    metrics['有效_比例'] = metrics['有效数量'] / filter_value.shape[0]

    # 发明授权和发明申请专利数量
    metrics['发明授权'] = (filter_value['专利类型'] == '发明授权').sum()
    metrics['发明申请'] = (filter_value['专利类型'] == '发明申请').sum()

    metrics['实用新型']=(filter_value['专利类型'] == '实用新型').sum()
    metrics['外观设计']=(filter_value['专利类型'] == '外观设计').sum()

    return metrics
```

```python
%%time
results = []

for m in range(219,655):#219,655
    co_month_t=co_month_all[co_month_all['月份数'] == m]
    patent_value=patent_co_1[patent_co_1['月份数'] <= m]
    
    group_ids_t=patent_value.groupby('股票代码')['公开（公告）号'].apply(list)
    ids_vec = group_ids_t.apply(lambda x: doc_lda_w2v.loc[doc_lda_w2v.index.isin(x)].mean(axis=0))    #ids_vec.index = group_ids_t['股票代码']
    ids_vec = ids_vec[ids_vec.sum(axis=1) != 0]#使用 lda_w2v 模型将公司的专利文档向量化后的表示
    if not ids_vec.empty:
        cosine_sim=pd.DataFrame(cosine_similarity(ids_vec.values), index=ids_vec.index, columns=ids_vec.index)

    ids_ipc = pd.crosstab(patent_value['股票代码'], patent_value['IPC主分类'].str.strip().str[:3])
    ids_ipc = ids_ipc[ids_ipc.sum(axis=1) != 0] #.reindex(columns=sorted(ipc_base.columns), fill_value=0)
    if not ids_ipc.empty:
        cosine_sim_ipc=pd.DataFrame(cosine_similarity(ids_ipc.values), index=ids_ipc.index, columns=ids_ipc.index)
  
    invest_sorted = co_month_t.dropna(subset='Mretwd_pre').sort_values(by=['证券代码', '月收盘日期']) 
    print(m,co_month_t['交易月份'].dropna().unique(),len(co_month_t),len(invest_sorted),len(group_ids_t),len(ids_vec),len(ids_ipc))
    #-------
    for index,row in co_month_t.iterrows():
        company_id = row['证券代码']
        last_date = row['月收盘日期']
        pre_month = row['Month_shift']

        if company_id in ids_vec.index: #company_id in cosine_sim.index 
            co_month_all.at[index,'listed_tech_closenesss'] = cosine_sim.loc[company_id].mean()
            recent_invest = invest_sorted[(invest_sorted['证券代码'].isin(cosine_sim.columns))&(invest_sorted['月收盘日期'] < last_date)].drop_duplicates(subset=['证券代码'], keep='last')
            if not recent_invest.empty:#recent_invest['Mretwd_pre'] notna
                tech_closeness_scores = cosine_sim.loc[company_id, recent_invest['证券代码']].fillna(0)  # 获取该公司与其他公司的技术相似度 
                weighted_vc_amount = np.dot(tech_closeness_scores.values, recent_invest['Mretwd_pre'].values) / tech_closeness_scores.sum()
                co_month_all.at[index,'tech_closeness_ret'] = weighted_vc_amount

        if company_id in  ids_ipc.index:
            co_month_all.at[index, 'listed_ipc_closenesss'] =  cosine_sim_ipc.loc[company_id].mean()
            recent_invest = invest_sorted[invest_sorted['证券代码'].isin(cosine_sim_ipc.columns)&(invest_sorted['月收盘日期'] < last_date)].drop_duplicates(subset=['证券代码'], keep='last')
            if not recent_invest.empty:
                tech_closeness_scores = cosine_sim_ipc.loc[company_id, recent_invest['证券代码']].fillna(0) 
                tech_closeness_sum = tech_closeness_scores.sum()
                if tech_closeness_sum != 0:
                    weighted_vc_amount = np.dot(tech_closeness_scores.values, recent_invest['Mretwd_pre'].values) / tech_closeness_sum
                    co_month_all.at[index,'ipc_closeness_ret']=weighted_vc_amount
                
         #-------
        filter_patent= patent_value[patent_value['股票代码']==company_id]#t <= m and company_id
        if not filter_patent.empty:
            filter_value = filter_patent if pd.isna(pre_month) else filter_patent[pre_month < filter_patent['月份数']]#trade_pre_t < t <= m and company_id,first:t <= m
            co_month_all.at[index,'知识宽度'] = process_ipc(filter_patent) # 累积量，fillna(method='ffill')
            if not filter_value.empty:
                if filter_value['月份数'].min()==m: # m in filter_value['月份数']
                    co_month_all.at[index,'m_type']|=(1<<1) #标记当期产生专利
            
                metrics = calculate_metrics(filter_value)   # 计算指标
                co_month_all.loc[index, metrics.keys()] = pd.Series(metrics)     # 批量更新
            #-------针对发明授权和发明申请类型的专利进行进一步处理
            mask=(filter_patent['专利类型']=='发明授权') | (filter_patent['专利类型']=='发明申请')
            if mask.sum()>0:
                filter_patent= filter_patent[mask]
                filter_value = filter_patent if pd.isna(pre_month) else filter_patent[pre_month < filter_patent['月份数']]
                metrics = {'知识宽度': process_ipc(filter_patent)}
                if not filter_value.empty:
                    if filter_value['月份数'].min()==m: 
                        metrics['m_type']= row['m_type']|(1<<1) 
                
                    metrics.update(calculate_metrics(filter_value))
                    
                results.append((index,metrics))
```
